```python
# Plotting a bar chart for Gender distribution
gender_counts = data['Traveler gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="pastel")
plt.title("Gender Distribution in Traveler Dataset")
plt.xlabel("Traveler gender")
plt.ylabel("Count")  # Corrected ylabel to "Count"
plt.savefig("gender_distribution.png")
plt.show()  # Show the plot
```

```python
# Load your dataset
data = pd.read_csv(r'C:\Users\SowmiS\Downloads\TravelerTripDataset\travel_details_dataset.csv')

# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

#correlation matrix
correlation_matrix = numeric_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig("Correlation_matrix.png")
plt.show()
```

```python
# Drop rows with missing in 'Accommodation cost' column
data.dropna(subset=['Accommodation cost'], inplace=True)

# Remove non-numeric characters from the 'Accommodation cost' column and converting to float
data['Accommodation cost'] = data['Accommodation cost'].replace('[^\d.]', '', regex=True).astype(float)

# Group by 'Accommodation type' and sum the costs
grouped_data = data.groupby('Accommodation type')['Accommodation cost'].sum().reset_index()

# Plotting the pie chart for Accommodation cost
plt.figure(figsize=(14, 14))
plt.pie(grouped_data['Accommodation cost'], labels=grouped_data['Accommodation type'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Accommodation Cost')
plt.savefig("Pie_chart.png")
plt.show()
```

```python
# Plotting a histogram
plt.figure(figsize=(8, 4))
plt.hist(data['Duration (days)'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Traveler Duration")
plt.xlabel("Duration (days)")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.savefig ("histogram_nationality.png")
plt.show()

```
